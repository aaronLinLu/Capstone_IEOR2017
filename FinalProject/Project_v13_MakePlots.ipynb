```python
import warnings
warnings.filterwarnings('ignore') # Filter out warnings
import numpy as np
import pandas as pd
import random as rnd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
from scipy import stats
import math
from math import isnan
from collections import namedtuple
from matplotlib.ticker import MaxNLocator
```

```python
df = pd.read_excel("Stats_predictionAccs.xlsx")
df.head(5)
```

```python
df.columns
```

```python
n_stocks = 3
plt.style.use("seaborn-white")

time_All = np.array(df['Processing Time, All '])
time_NoCancel = np.array(df['Processing Time, Exclude Cancellation'])

fig, ax = plt.subplots(1, 1, figsize=(8, 6))

index = np.arange(n_stocks)
bar_width = 0.15

opacity = 1.0
# error_config = {'ecolor':'0.3'}

rects1 = ax.bar(index,time_All,bar_width,alpha=opacity,color='gold',
               label='Processing Time, All Messages')
rects2 = ax.bar(index+bar_width,time_NoCancel,bar_width,alpha=opacity,color='darkblue',
               label='Processing Time, Exclude Cancellation')

ax.set_xlabel('Stock')
ax.set_ylabel('Average Processing Time In Seconds')
ax.set_title('Processing Time Comparison')
ax.set_xticks(index + bar_width/2)
ax.set_xticklabels(('Amazon','Apple','Google'))
ax.legend()

fig.tight_layout()
plt.show()
```

```python
n_stocks = 3
plt.style.use("seaborn-white")

accs_All = np.array(df['Accuracy, All'])
accs_NoCancel = np.array(df['Accuracy, Exclude Cancellation'])

fig, ax = plt.subplots(1, 1, figsize=(8, 6))

index = np.arange(n_stocks)
bar_width = 0.15

opacity = 1.0
# error_config = {'ecolor':'0.3'}

rects1 = ax.bar(index,accs_All,bar_width,alpha=opacity,color='gold',
               label='Average Accuracy, All Messages')
rects2 = ax.bar(index+bar_width,accs_NoCancel,bar_width,alpha=opacity,color='darkblue',
               label='Average Accuracy, Exclude Cancellation')

ax.set_xlabel('Stock')
ax.set_ylabel('Average Prediction Accuracy')
ax.set_title('Average Accuracy Comparison')
ax.set_xticks(index + bar_width/2)
ax.set_xticklabels(('Amazon','Apple','Google'))
ax.legend()

fig.tight_layout()
plt.show()
```
