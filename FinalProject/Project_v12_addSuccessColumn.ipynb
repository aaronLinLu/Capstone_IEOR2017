```python
import pandas as pd
```

```python
df = pd.read_csv("AMZN.csv")
```

```python
df.head(5)
```

```python
df.columns
```

```python
success = []

for index, row in df.iterrows():
    if (row['Label_st']== 'la' and row['Level_ed']<= row['Level_st'] 
    and row['mid-price 1_ed'] <= row['mid-price 1_st']):
        success.append(True)
    elif (row['Label_st']== 'lb' and row['Level_ed']<= row['Level_st'] 
    and row['mid-price 1_ed'] >= row['mid-price 1_st']):
        success.append(True)
    else:
        success.append(False)
print(success[:50])
```

```python
df['Success'] = success
```
